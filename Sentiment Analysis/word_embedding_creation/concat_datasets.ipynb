```
import pandas as pd
```

# Load Data

```
folder_path = '/content/drive/My Drive/University/FYP/Sentiment Analysis/Implementation/'

lankadeepa_tagged_path = folder_path + 'corpus/newly_tagged/Lankadeepa.csv' #isuru preprocessing
gossiplanka_tagged_path = folder_path + 'corpus/newly_tagged/Gossip_Lanka.csv' 
tagged_concated_dfs_path = folder_path + 'corpus/newly_tagged/Lankadeepa_and_GossipLanka.csv'

lankadeepa_all_path = folder_path + 'corpus/analyzed/lankadeepa_all_2.csv' #isuru preprocessing
gossiplanka_all_path = folder_path + 'corpus/GossipLanka_Comments_all.csv' #
```

# Concat Dataframes Tagged

```
Lankadeepa = pd.read_csv(lankadeepa_tagged_path)
Gossip_Lanka = pd.read_csv(gossiplanka_tagged_path)

Lankadeepa = Lankadeepa.rename(columns={"corrected_label":"label"})
Gossip_Lanka = Gossip_Lanka.drop(columns="Unnamed: 3")

concated_dfs = pd.concat([Lankadeepa,Gossip_Lanka])
concated_dfs.to_csv(tagged_concated_dfs_path)
```
